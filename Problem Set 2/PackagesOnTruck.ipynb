```python
max_weight = float(input("Please enter the weight limit of the truck: "))
print()
in_truck = 0

packages_north = 0
packages_south = 0
packages_east = 0
packages_west = 0

count_packages = 0

while in_truck < max_weight:
    pkg_weight = float(input("Please enter the weight of a package: "))
    in_truck += pkg_weight
    if in_truck > max_weight:
        break
    
    pkg_no = int(input("Please enter the package ID number: "))
    pkg_zone = input("Please enter the zone: ")
    count_packages += 1
    print()
    print("The following package is on the truck:")
    print("Pkg No. = {} , Zone = {} , Pkg Wgt = {}" .format(pkg_no, pkg_zone, pkg_weight))
    print()

    
    if pkg_zone == "North":
        packages_north += 1
    elif pkg_zone == "South":
        packages_south += 1
    elif pkg_zone == "East":
        packages_east += 1
    elif pkg_zone == "West":
        packages_west += 1
    
print()                
print ("{} packages are on the truck for North zone" .format(packages_north))
print ("{} packages are on the truck for South zone" .format(packages_south))
print ("{} packages are on the truck for East zone" .format(packages_east))
print ("{} packages are on the truck for West zone" .format(packages_west))
print()
print("{} packages are loaded on the truck" .format(count_packages))
            
        
        
        
    
    
```
